```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv(r'E:\python\代码存储\爬虫代码\requests库\douban_top250.csv')
df
```

```python
print(df.isnull().sum())
print(df.info())
```

```python
data = df['评分']
plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'Arial']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize = (10 , 6))
data.hist(bins = 10 , color = 'blue')
plt.title("豆瓣top250电影评分分布")
plt.xlabel("电影数量")
plt.ylabel("电影评分")
plt.savefig("rating_hist.png")
plt.show()
```

```python
genre_num = df['类型'].str.split(",").explode().value_counts()
genre_num
```

```python
plt.figure(figsize = (20,5))
plt.plot(genre_num, marker='o', linestyle='-', color='b', label='电影类型')
plt.savefig("genre_num_plot.png")
plt.title("豆瓣top250电影类型统计")
plt.xlabel("电影类型")
plt.ylabel("类型数量")
plt.show()
```
